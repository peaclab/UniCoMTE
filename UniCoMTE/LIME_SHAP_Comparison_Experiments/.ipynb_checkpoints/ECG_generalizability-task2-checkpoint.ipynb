# Are the explanations generalizable among other test instances?

* Question: How many instances does an explanation cover? Is one explanation applicable to many test cases?

* Use HPC Time series data
* Use existing pipeline and existing explainability framework

* Experiment 1:

    * For each explanation, perform the feature substitution to every other sample that matches (true class, mispredicted class)
    * Record how many other samples of the same class flip their prediction
    * Report the coverage

* Experiment 2: 

    * Get explanations for each sample in test set for an (or all) opposing class(es)
    * For each explanation, perform the same substitutions to every other sample with a matching class
    * Record how many other samples flip their prediction
    * Report the coverage

```python
def dict_to_pickle(dict,filename):

    dir = "/projectnb/peaclab-mon/aksar/explainability_experiments/generality/"
    with open(dir+filename+'.pickle', 'wb') as f:
        pickle.dump(dict,f)
    
def pickle_to_dict(filename):
    
    dir = "/projectnb/peaclab-mon/aksar/explainability_experiments/generality/"
    with open(dir+filename+'.pickle', 'rb') as f:
        loaded_pickle = pickle.load(f)
    
    return loaded_pickle
```

```python
#This only support with one column dataframes
def plot_utils(scores,key_name,title,hist=False,col_name=0):
    df_utils = convert_df(scores,key_name)
    plot_me_before_you_go(df_utils,title,hist=hist,col_name=key_name)    
```

```python
def convert_df(scores,key_name):
    score_coverage = []
    for ind in range(len(scores)):
        score_coverage.append(scores[ind][key_name])
    
    if type(score_coverage[0]) == float or type(score_coverage[0]) == int :
        df_coverage= pd.DataFrame(score_coverage,columns=[key_name])
        print(df_coverage[key_name].mean())
    else:    
        df_coverage= pd.DataFrame(score_coverage)
    
    return df_coverage
```

```python
#Will plot coverage histogram
def plot_me_before_you_go(data,title,directory=None,hist=False,col_name=0):

    if hist:
        
        fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(10, 4))
        graph = sns.distplot(data[col_name],ax=ax1,kde=False,bins=15)

        ax1.set_xlabel('Explanation index')
        ax1.set_title(title)
        plt.tight_layout()

        if directory is not None:
            plt.savefig(directory)        
    
    else:
        fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(10, 4))
        graph = sns.barplot(x=data.index, y=data[col_name],ax=ax1,
                  palette=palette,
                   capsize=0.07, errwidth=2, errcolor='#444444', edgecolor='#666666')


        graph.axhline(data[col_name].mean(),c='r',linewidth=1.8)

        ax1.set_xlabel('Explanation index')
        ax1.set_title(title)
        plt.tight_layout()

        if directory is not None:
            plt.savefig(directory)        
```

```python
def run_generality(pipeline, timeseries, labels, true_label, pred_label,picklename):

    preds = pipeline.predict(timeseries)
    
    l = list(enumerate(zip(labels['label'], preds)))
    exp_set = {}
    
    for idx, (true, pred) in l:    
        if true != pred and true == true_label and pred == pred_label:
            x_test_idx = labels.iloc[idx].name[0]
            x_test = timeseries.loc[[x_test_idx], :, :]
            exp_set[idx] = x_test    

    ourmethod = explainers.OptimizedSearch(pipeline, timeseries,labels,num_distractors=1,threads=multiprocessing.cpu_count())
    columns = timeseries.columns
    scores = []
    to_maximize = np.argmax(pipeline.classes_ == true_label)

    logging.info("Explanation set length %d",len(exp_set))
    
    for idx, x_test in exp_set.items():
        
        baseline_exp, best_dist = ourmethod.explain(x_test,to_maximize=to_maximize,return_dist=True)
        counter = 0
        prob_list = []
        
        for idx,temp_test in exp_set.items():
            

            if temp_test.index.get_level_values("node_id").unique() != x_test.index.get_level_values("node_id").unique():      
                
                modified = temp_test.copy()
                prob_first = pipeline.predict_proba(modified)[0][to_maximize]                    
                
                for c in columns:
                    if c in baseline_exp:
                        modified[c] = best_dist[c].values
                    
                    
                prob_last = pipeline.predict_proba(modified)[0][to_maximize]
                prob_list.append(prob_last - prob_first)
                                
                if pipeline.predict(modified) == true_label:
                    counter += 1    
                    
            else:
                modified = temp_test.copy()
                prob_first = pipeline.predict_proba(modified)[0][to_maximize]                    
                
                for c in columns:
                    if c in baseline_exp:
                        modified[c] = best_dist[c].values
                    
                    
                prob_last = pipeline.predict_proba(modified)[0][to_maximize]
                
                if np.argmax(pipeline.predict_proba(modified)[0]) != to_maximize:
                    print(prob_first)
                    print("##########")
                    print(prob_last)
                                        
                prob_xtest = prob_last - prob_first
                print(prob_xtest)
                                                
                                
        coverage = counter / (len(exp_set)-1)
        scores.append({
                'explanation': baseline_exp,
                'count': counter,
                'coverage': coverage,
                'prob_xtest': prob_xtest,
                'prob_change': prob_list,
                'x_test_node_id': x_test.index.get_level_values('node_id').unique().values[0],                                
                'dist_node_id':best_dist.index.get_level_values('node_id').unique().values[0]                                
        })    
        logging.info("Explanation: %s",baseline_exp)
        logging.info("Coverage: %f" ,coverage)
        logging.info("Prob list: %s",prob_list)
        logging.info("node id %s",scores[-1]['dist_node_id'])
        
    dict_to_pickle(scores,picklename)
        
    return scores 
```
