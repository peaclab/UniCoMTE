```python
import numpy as np
import pandas as pd
import sys
sys.path.insert(0,"/projectnb/peaclab-mon/JLi/projectx/CoMTE_V2_JLi/ECG_Visualization/ecg_plot_counterfactual")
import ecg_plot_counterfactual 
```

```python
# load in KDTree + inspect
import pickle

kd_tree_path = "/projectnb/peaclab-mon/JLi/projectx/AutomaticECGDiagnosis/KDTree_all_training/combined_kdtrees.pkl"

with open(kd_tree_path, 'rb') as f:
    print('loading kdtree from pkl')
    combined_kdtrees = pickle.load(f)

for key, kdtree in combined_kdtrees.items():
    print(f"Key: {key}")
    print("Data points in the KDTree:")
    print(kdtree.data.shape)  # Print the data points stored in the KDTree
    print("-" * 50)  # Print a separator line for clarity
```

```python
# analyze flatline vs. non-flatlined samples
flatline_idx = {241987,
                191664,
                173587,
                16276,
                2503}
                # 128186,
                # 200550,
                # 105241,
                # 199739,
                # 167677,
                # 101082,
                # 211539,
                # 105506,
                # 152205,
                # 61685,
                # 221222,
                # 92533,
                # 199794,
                # 142860,
                # 77362,
                # 147504,
                # 267315,
                # 156202,
                # 222377,
                # 65574,
                # 49150,
                # 122168,
                # 77297,
                # 98568,
                # 173823,
                # 118504,
                # 255993,
                # 152912,
                # 162368}

regular_idx = {1,2,3,4,5}

samples = combined_kdtrees[0].data
print(type(samples))
```

```python
# analyze flatline samples

# use nested dictionary to store statistics
feature_dict = {}

# calc metrics
for idx in flatline_idx:
    # when there are no self.per_class_node_indices (since KDTree was given)
    # if not self.per_class_node_indices:
    flat = np.array(samples[idx])
    sliced_distractor = flat.reshape(12, 4096).T
    # set distractor columns
    column_names = ['DI', 'DII', 'DIII', 'AVR', 'AVL', 'AVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
    # Create the DataFrame with the given column names
    sliced_distractor = pd.DataFrame(sliced_distractor, columns=column_names)


    # visualization
    distractor_visualization = sliced_distractor.to_numpy().T        
    # plot distractor returned
    save_path = f"/projectnb/peaclab-mon/JLi/projectx/CoMTE_V2_JLi/ECG_utils/ECGCounterfactual_sample{idx}.jpg"
    ecg_plot_counterfactual.plot(distractor_visualization, distractor_visualization, [], save_path, sample_rate = 400, title = 'Standard 12 Lead Electrocardiogram Recording', style = 'bw', row_height = 10)


    # analyze features
    minimum = np.min(distractor_visualization)
    maximum = np.max(distractor_visualization)
    std_dev = np.std(distractor_visualization)
    feature_dict[idx] = {
        'min': minimum,
        'max': maximum,
        'std_dev': std_dev
    }

print(feature_dict)
```

```python
# analyze regular samples

# use nested dictionary to store statistics
feature_dict = {}

# calc metrics
for idx in regular_idx:
    # when there are no self.per_class_node_indices (since KDTree was given)
    # if not self.per_class_node_indices:
    flat = np.array(samples[idx])
    sliced_distractor = flat.reshape(12, 4096).T
    # set distractor columns
    column_names = ['DI', 'DII', 'DIII', 'AVR', 'AVL', 'AVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
    # Create the DataFrame with the given column names
    sliced_distractor = pd.DataFrame(sliced_distractor, columns=column_names)


    # visualization
    distractor_visualization = sliced_distractor.to_numpy().T        
    # plot distractor returned
    save_path = f"/projectnb/peaclab-mon/JLi/projectx/CoMTE_V2_JLi/ECG_utils/ECGCounterfactual_sample{idx}.jpg"
    ecg_plot_counterfactual.plot(distractor_visualization, distractor_visualization, [], save_path, sample_rate = 400, title = 'Standard 12 Lead Electrocardiogram Recording', style = 'bw', row_height = 10)


    # analyze features
    minimum = np.min(distractor_visualization)
    maximum = np.max(distractor_visualization)
    std_dev = np.std(distractor_visualization)
    feature_dict[idx] = {
        'min': minimum,
        'max': maximum,
        'std_dev': std_dev
    }

print(feature_dict)
```

```python
# filter with vectorized method (chunk)
from scipy.spatial import KDTree

batch_size = 280394

batch = np.array(samples[:batch_size])
print(batch.shape)

std_devs = np.std(batch, axis=1)
print(std_devs.shape)

threshold = 0.1  # Adjust as needed
mask = std_devs >= threshold
# print(mask)
print(mask.shape)


# Filter out low-std-dev (flatline) samples
filtered_samples = batch[mask]
print(filtered_samples.shape)

# log indices
removed_indices = np.where(~mask)[0]
kept_indices = np.where(mask)[0]

# print(removed_indices)
# print(kept_indices)

# turn into kdTree
new_tree = KDTree(filtered_samples)
```

```python
# save off new kd tree
import pickle

file_path = "/projectnb/peaclab-mon/JLi/projectx/AutomaticECGDiagnosis/KDTree_all_training/filtered_0_kd_tree.pkl"

with open(file_path, 'wb') as file:
    pickle.dump(new_tree, file)

```
